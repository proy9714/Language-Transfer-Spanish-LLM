```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
filename = "assets/Complete+Spanish+transcript+-+2019+final.pdf"
```

```python
loader = PyPDFLoader(filename)
pages = loader.load_and_split()
```

```python
# Separate and extract all the Tracks

tracks = []
track_content = ("").join([page.page_content for page in pages[7:10]])
for page in pages[10:]:
    if page.page_content.startswith("Track"):
        tracks.append(track_content)
        track_content = page.page_content
        
    else:
        track_content += page.page_content 
        
tracks.append(track_content)
```

```python
tracks
```

```python
# There should be exactly 90 tracks like in the spanish language transfer course

len(tracks)
```

```python
# Save these separate tracks as pdf files

parent_file_path = "assets/Tracks"
for index, track in enumerate(tracks):
    with open(f"{parent_file_path}/{index+1}.txt", "w") as file:
        file.write(track)
```

```python
# Modify the track sources

track_number = 0
for index, page in enumerate(pages[7:]):
    if page.page_content.startswith("Track"):
        track_number += 1
        
    page.metadata["source"] = f"Track {track_number}"
```

```python
# To be used from this notebook

from langchain_community.document_loaders import PyPDFLoader

class SourceDocumentLoader:
    
    def __init__(self, filepath):
        self.filepath = filepath
        self.pages = None
        
    # Load the pdf
    def load_pdf():
        loader = PyPDFLoader(self.filepath)
        self.pages = loader.load_and_split()
        
        print(f"Loaded {len(pages)} pages!")
        
        self.modify_sources()
        
    # Modify the track sources
    def modify_sources():
        track_number = 0
        for index, page in enumerate(self.pages[7:]):
            if page.page_content.startswith("Track"):
                track_number += 1
                
            page.metadata["source"] = f"Track {track_number}"
            
    def get_pages():
        return self.pages
```
